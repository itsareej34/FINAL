```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
df = pd.read_csv("/content/quarter-.csv")
```

```python
df.head()
```

```python
plt.scatter(df.yearnumber,df.quarterid,marker='*',color='blue')
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df[['yearnumber']],df.quarterid,train_size=0.8)
X_test
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train, y_train)
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train, y_train)
X_test
```

```python
y_predicted = model.predict(X_test)
model.predict_proba(X_test)
model.score(X_test,y_test) # Accuracy of the model
y_predicted
X_test
```

```python
#model.coef_ indicates value of m in y=m*x + b equation
model.coef_
#model.intercept_ indicates value of b in y=m*x + b equation
model.intercept_
```

```python
import math
def sigmoid(x):
  return 1 / (1 + math.exp(-x))
def prediction_function(yearnumber):
    z = 0.042 * yearnumber - 1.53
    y = sigmoid(z)
    return y
```
